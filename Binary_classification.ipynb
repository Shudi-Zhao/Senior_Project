```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
import pickle

tfidf = TfidfVectorizer(ngram_range=(1,2),norm='l2', stop_words = stop_words, max_df = 0.4, min_df = 0.01)

LR = LogisticRegression(random_state = 1, class_weight='balanced', multi_class = 'ovr', penalty='l1', C= 1, solver= 'liblinear', max_iter = 300)


pipe = Pipeline([('tfidf', tfidf),
                 ('LogisticRegression',LR)
                 ])
```

```python
# Create a function to do prediction based on given threshold or probability for education
def my_prediction(Xtest, threshold = 0.5): #threshold for educational
    pred_prob = model.predict_proba(Xtest)

    pred = []
    for i in pred_prob[:, 0]: #the probability of being educational contents
        if i > threshold:
            pred.append('edu')
        else:
            pred.append('nonedu')
    preds = np.array(pred)
    return preds
```

```python
import time
start = time.time()

model = pipe.fit(Xtrain,ytrain)
```

```python
preds = my_prediction(Xtest, 0.1)
preds2 = my_prediction(Xvalid, 0.1)
```

```python
# save model
filename = 'LR_Binary.sav'
pickle.dump(model, open(filename, 'wb'))
```

# SVM

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
import pickle

tfidf = TfidfVectorizer(ngram_range=(1,2),norm='l2', stop_words = stop_words, max_df = 0.4, min_df = 0.01)

svc = SVC(kernel = 'linear', class_weight = 'balanced')

pipe = Pipeline([
    ('tfidf', tfidf),
    ('clf', svc)
])
```

```python
import time
start = time.time()

model = pipe.fit(Xtrain,ytrain)
preds = model.predict(Xtest)
```

```python
filename = 'SVM_Binary.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, preds, digits = 4))
```

```python
# Classificaiton_report for validation set
preds2 = model.predict(Xvalid)
print(classification_report(yvalid, preds2, digits = 4))
```

```python
print("--- Executed in %s minutes ---" % round((time.time() - start)/60, 3))
```

# Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
import pickle

tfidf = TfidfVectorizer(ngram_range=(1,2),norm='l2', stop_words = stop_words, max_df = 0.4, min_df = 0.01)
Naive_bayes = MultinomialNB()

pipe = Pipeline([
    ('tfidf', tfidf),
    ('MultinomialNB', Naive_bayes)
])
```

```python
import time
start = time.time()

model = pipe.fit(Xtrain,ytrain)
preds = model.predict(Xtest)
```

```python
filename = 'NB_Binary.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
from sklearn.metrics import classification_report
# Classification_report for test set
print(classification_report(ytest, preds, digits = 4))
```

```python
# Classificaiton_report for validation set
preds2 = pipe.predict(Xvalid)
print(classification_report(yvalid, preds2, digits = 4))
```

```python
print("--- Executed in %s minutes ---" % round((time.time() - start)/60, 3))
```
